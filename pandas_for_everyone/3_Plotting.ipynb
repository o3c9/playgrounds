```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
ds = sns.load_dataset("anscombe")
```

```python
ds_1 = ds[ds["dataset"] == "I"]
plt.plot(ds_1["x"], ds_1["y"])
```

```python
plt.plot(ds_1["x"], ds_1["y"], "o")
```

```python
ds_2 = ds[ds["dataset"] == "II"]
ds_3 = ds[ds["dataset"] == "III"]
ds_4 = ds[ds["dataset"] == "IV"]

fig = plt.figure(figsize=(12,8))

axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

axes1.plot(ds_1["x"], ds_1["y"], "o")
axes1.set_title("dataset I")
axes2.plot(ds_2["x"], ds_2["y"], "o")
axes2.set_title("dataset II")
axes3.plot(ds_3["x"], ds_3["y"], "o")
axes3.set_title("dataset III")
axes4.plot(ds_4["x"], ds_4["y"], "o")
axes4.set_title("dataset IV")

fig.tight_layout()
```

```python
# Multivariable Plotting

tips = sns.load_dataset("tips")
```

```python
def recode_sex(val):
    if val == "Female":
        return 1
    else:
        return 0
    
tips["sex_color"] = tips["sex"].apply(recode_sex)
scatter_plot = plt.figure(figsize=(8, 6))
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(x=tips["total_bill"], y=tips["tip"], s=tips["size"]* 10, c=tips["sex_color"], alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.show()
```
