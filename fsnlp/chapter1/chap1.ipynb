### 1.3

ランダムな文字列生成器を用いて、ジップの法則の確認する

```python
import random

def generate_document(n = 1000000):
    chars = list(map(chr, range(97, 123))) + [" "] * 8
    generate = lambda: random.choice(chars)
    return "".join([generate() for i in range(n)])

doc = generate_document()
```

```python
import pandas as pd
words = doc.split(" ")

sf = pd.Series(words).value_counts()
```

```python
sff = sf.value_counts()
sff
```

```python
import numpy as np
sfff = pd.DataFrame([sff.index, sff.values, np.multiply(sff.values, sff.index)]).T
sfff
```

### 1.4 連語を見つける

テキストには、ジェーン・オースティンの「説き伏せられて」（説得）を用いる

#### まずは単純なbi-gramを見てみよう

```python
from nltk.tokenize import RegexpTokenizer

text = nltk.corpus.gutenberg.raw("austen-persuasion.txt")
sentences = " ".join(text.split("\n"))
tokenizer = RegexpTokenizer(r'\w+')
words = tokenizer.tokenize(sentences)
```

```python
list = []
for i in range(len(words)):
    if i == 0:
        list.append(["<unk>", words[i]])
    elif i == len(words):
        list.append([words[i], "<unk>"])
    else:
        list.append([words[i-1], words[i]])

bigrams = [" ".join(e) for e in list]
```

```python
pd.Series(bigrams).value_counts()
```

#### ここでは一例としてstopwordsを除去してみる

```python
# nltk.download('stopwords')
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
non_stop_list = []

for pair in list:
    if pair[0] not in stops and pair[1] not in stops:
        non_stop_list.append(" ".join(pair))

pd.Series(non_stop_list).value_counts()
```
